Given the time marching scheme:
$$u_{n+1} = u_{n-1} + \cfrac{2}{3}h\left(f_{n+1}+f_n+f_{n-1}\right)$$

### Part a

Write the O$\Delta$E for the representative equation.

with $f(u) = \lambda u$, the scheme is

$$u_{n+1} = u_{n-1} + \cfrac{2}{3}h\lambda\left(u_{n+1}+u_n+u_{n-1}\right)$$

### Part b

The amplification factor, $\sigma$ is calculated to determine stability of the method.

With the general form of

$$u_{n+k} = \sigma^k u_n$$

we can turn the time marching scheme into:

$$\begin{align}
\sigma u_n = \cfrac{1}{\sigma}u_n + \cfrac{2}{3}h\lambda\left(\sigma u_n+ u_n+\cfrac{1}{\sigma}u_n\right)
\end{align}$$

Which algebras into

$$\begin{align}
\sigma^2  &= 1 + \alpha\sigma^2 +\alpha\sigma +\alpha\\
(1-\alpha)\sigma^2 - \alpha\sigma - (1+\alpha) & = 0
\end{align}$$

where $\alpha = \frac{2}{3}\lambda h$

Using the quadratic formula, we find the roots

$$\begin{align}
\sigma_{1,2} & = \cfrac{-\alpha\pm\sqrt{4-3\alpha^2}}{2(\alpha-1)}
\end{align}$$

The principal root comes out of the (-) of the $\pm$ above, we determine this by setting $\alpha$ (i.e. $\lambda h$) equal to zero. Then the root giving $\sigma = 1$ is the principal root we desire.

So let's write this as

$$\sigma_1 =  \cfrac{-\alpha-\sqrt{4-3\alpha^2}}{2(\alpha-1)}$$

### Part c

I'll make a mesh grid of 'x' and 'y' for the real and imaginary axis. Then I can simply 

```python
x = linspace(-1.,1.,101)
y = linspace(-1.,1.,101)
xx, yy = meshgrid(x,y)
A = 2/3.*(xx + 1j*yy)

#sigma1 = (-A + sqrt(4. - 3*A**2.))/(2*(A-1))
#sigma1 = sigma1*where(absolute(sigma1)<1,1,0)
sigma2 = (-A - sqrt(4. - 3*A**2.))/(2*(A-1))
sigma2 = sigma2*where(absolute(sigma2)<1,1,0) # wipe out sigma>1
```

In the above code, I have wiped out $\sigma_1$ anywhere that it is greater than 1. So now I can just contour the $\sigma_1$ remaining and it will show where it's stable.

```python
levels = linspace(-20,1,1001)

contour(xx,yy,sigma2,levels)
```

and we see that it's A-stable because the stability region is the entire LHS of the complex plane.

### Part d

```python
from mpl_toolkits.mplot3d import Axes3D

p=3
f = (exp(A)-sigma2)/A**p
#contour(xx,yy,f)
fig = figure(str(p))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(xx, yy, f, rstride=1, cstride=1, cmap=cm.coolwarm,
        linewidth=0, antialiased=False)
show()
```
