We are to build a Taylor Table to construct a one-sided compact finite difference scheme to approximate the first derivative with a stencil that includes derivatives at j and j+1 and values of the function at j, j + 1 and j + 2. We are then to show that the scheme is 3rd-order-accurate.

### Part A
The Taylor Table's construction is shown in the accompanying Word document. From that table, we construct a system of equations

$$\begin{align}
\alpha_0 + \alpha_1 + \alpha_2 & = 0 \\
-\beta_1 + \alpha_1 + 2\alpha_2 & = 1 \\
-2\beta_1 + \alpha_1 + 2^2\alpha_2 & = 0 \\
-3\beta_1 + \alpha_1 + 2^3\alpha_2 & = 0
\end{align}$$

which gives us 4 equations for 4 of the unknown coefficients. We're left with an $O(\Delta x)^4$ equation that can not be satisfied. Thus we write

$$\Delta x u_j' \beta_1\Delta x u_{j+1}' = \alpha_0 u_j  + \alpha_1 u_{j+1} + \alpha_2 u_{j+2} + O(\Delta x^4)$$

or,

$$u_j'  = -\beta_1 u_{j+1}' +\cfrac{1}{\Delta x}\left( \alpha_0 u_j  + \alpha_1 u_{j+1} + \alpha_2 u_{j+2}\right) + O(\Delta x^3)$$

and we see that with this method, to find the first derivative at $u_j$, we will be 3rd-order-accurate.

The LHS of the system of equations are put into the form of a matrix, $A$. The RHS are in the array, $f$. Solving for the unknown coefficients (an array $x$ here) gives:

```python
A = array([[1., 1., 1., 0.],[0., 1., 2., -1.],[0., 1., 4., -2.],[0., 1., 8., -3.]])
B = linalg.inv(A)
f = array([[0.],[1.],[0.],[0.]])
x = dot(B,f)
print A
print x
```

First, the derivative

$$\begin{align}

$$\begin{align}

$$\begin{align}

#### 4th-order compact centered difference scheme

Using a Taylor Table from class notes, we extract coefficients for the 4th-order compact centered difference scheme. Thus the 4-th order compact centered difference scheme stencil is

$$u_{i-1}' + 4u_i'+ u_{i+1}' = \cfrac{3}{\Delta x}\left(-u_{i-1} + u_{i+1}\right)$$

To address the boundary conditions, let's look at $i = 0$. 

$$u_{N-1}' + 4u_0'+ u_{1}' = \cfrac{3}{\Delta x}\left(-u_{N-1} + u_{1}\right)$$

When $i = N$, we have the same equation. Thus we only need to run up to $N-1$ in our array and we have a circulant tridiagonal matrix on the LHS, $B_p(N-1,1,4,1)$. We'll use our trisolve to solve this equation.

I don't want to import the trisolve as an external library. I know all library dependencies (numpy, scipy, matplotlib, etc.) are loaded when I load IPython, but have to call them separately if it were an external Python file. So I just paste this code from the first homework problem. I will have to clean this up for later work...

```python
def trisolve(a, b, c, f, solveType):
    M = len(f);
    if solveType == 'circ':
        if isscalar(a) == True and isscalar(b) == True and isscalar(c) == True:
            m = arange(0,M)
            lam = b + (a+c)*cos(2*pi*m/M) - 1j*(a-c)*sin(2*pi*m/M);
            fhat = fft.fft((f))/M
            xhat = fhat/lam 
            x = M*fft.ifft(xhat)
            return x.real
        else:
            print "Error: circulant matrix solver requires scalar inputs"
    elif solveType == 'reg':
        if isscalar(a) == True:
            a = ones(M)*a
        if isscalar(b) == True:
            b = ones(M)*b
        if isscalar(c) == True:
            c = ones(M)*c
        for i in xrange(M-1):
            f[i+1] -= f[i] * a[i] / b[i]
            b[i+1] -= c[i] * a[i] / b[i]
        for i in reversed(xrange(M-1)):
            f[i] -= f[i+1] * c[i] / b[i+1]
        return [f[i] / b[i] for i in xrange(M)]
```

#### 2nd-order explicit central difference scheme

We can do it again for an explicit central difference scheme.  The stencil for this is

$$u_i' \approx \cfrac{u_{i+1} - u_{i-1}}{2\Delta x}$$

Again, because of periodicity, we need not worry about the $\pm$ nodes at the boundaries. We can solve for the derivative for a range of M

```python
from scipy.linalg import circulant

# the exact solution
x2 = linspace(0.,1,1001)
u2 = 2*pi*cos(2*pi*x2)

# entries for the ciruclant matrix
a = 0.
c = 0.
b = 1.
l2_err2 = []

m = [2**i for i in range(2,8)]

for M in m:
    temp_exact = zeros(M)
    f = zeros(M)
    
    #construct the RHS
    for i in range(M):
        f[i] = 1/(2*(1./M))*(-sin(2*pi*((i-1)*(1./M)))+sin(2*pi*((i+1)*(1./M))))
        temp_exact[i] = 2*pi*cos(2*pi*i*(1./M))
    
    # solve for the unknowns    
    y = trisolve(a,b,c,f,'circ')
    l2_err2 = append(l2_err2, linalg.norm(y-temp_exact))
    # because of periodicity, we only go up to the next-to-last point
    # but y(0) = y(N) so I'll append this for the plot
    y = append(y,y[0])
       
    # append the first location to the last to account for periodicity in the plot
    # and make a new x span over the M+1 locations.
    xPlot = linspace(0,1,M+1)
    figure(1)
    plot(xPlot,y)
    
figure(1)
plot(x2,u2,'k--')
legend(["M = " + str(i) for i in m],4)
title('Explicit, 2nd order, central difference scheme')
```

### Divergence

```python
def div( qx, qy ):
    # takes a function defined on the edges and returns the divergence
    # at the cell centers
    # input (M+1 ,M), (M, M+1)
    # output (M, M)
    
    return diff(qx,axis=0) + diff(qy,axis=1)
```

### Gradient

```python
def grad( fg ):
    # takes a function defined on cell centers and returns functions 
    # on the cell edges, the x edge first and y edge second
    # input (M ,M)
    # output (M+1, M), (M, M+1)

    gradFx = concatenate((array([fg[0,:]]),diff(fg,n=1,axis=0)\
                          ,array([-fg[-1,:]])),0)
    
    gradFy = concatenate((transpose(array([fg[:,0]])),diff(fg,n=1,axis=1)\
                          ,transpose(array([-fg[:,-1]]))),1)
        
    return gradFx, gradFy
```

The gradient function is evaluated at parameters located at 'centers'. It gets mapped to the 'edges'.

```python
gfx, gfy = grad( fc )
gfx = gfx/M
gfy = gfy/M
```

### Edge curl

```python
def edgecurl( qx, qy ):
    # takes a function defined at the edges and returns one defined
    # on the nodes
    # input (M+1 ,M), (M, M+1)
    # output (M+1, M+1)

    A = concatenate((array([qy[0,:]]),diff(qy,n=1,axis=0)\
                     ,array([-qy[-1,:]])),0)
    
    B = concatenate((transpose(array([qx[:,0]])),diff(qx,n=1,axis=1)\
                     ,transpose(array([-qx[:,-1]]))),1)
    
    return A-B
```
